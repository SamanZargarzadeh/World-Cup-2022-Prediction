# Box plot

```python
data_difs = Dataset[['h_rating',
 'h_eoy_rank',
 'h_boy_rank',
 'a_rating',
 'a_eoy_rank',
 'a_boy_rank']]
scaled = (data_difs - data_difs.mean()) / data_difs.std()
scaled["results"] = Dataset["results"]
violin = pd.melt(scaled,id_vars="results", var_name="features", value_name="value")

```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x="features", y="value", hue="results", data=violin , color='c')
plt.xticks(rotation=90)
plt.show()
```
