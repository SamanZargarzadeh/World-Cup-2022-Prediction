```python
import pandas as pd
```

```python
Matches=pd.read_csv(r'/Users/macbook/Desktop/final_dataset.csv')
Matches.head()
```

```python
hdf=Matches.groupby(['year','hname']).sum().reset_index()
hdf.head()
```

```python
adf=Matches.groupby(['year','aname']).sum().reset_index()
adf.head()
adf.columns
```

```python
adf2 = adf.set_axis(['year', 'hname', 'ID', 'agoals', 'hgoals', 'aPossesion', 'hPossesion',
       'ashotsOnTarget', 'hshotsOnTarget', 'ashots', 'hshots', 'ayellowCards',
       'hyellowCards', 'aredCards', 'hredCards', 'afouls', 'hfouls', 'asaves',
       'hsaves', 'a_rating', 'a_eoy_rank', 'a_boy_rank', 'h_rating',
       'h_eoy_rank', 'h_boy_rank'], axis=1, inplace=False)
adf2
```

```python
All=pd.concat([hdf, adf2] , axis=0, ignore_index=True)
Merged=All.groupby(['year','hname']).sum().reset_index()
Merged.head()

```

```python
hc=Matches.groupby(['year','hname']).count().reset_index()
hGames1=hc[["year","hname","ID"]]
hGames=hGames1.set_axis(["year","hname","Count"], axis=1, inplace=False)
```

```python
ac=Matches.groupby(['year','aname']).count().reset_index()
aGames1=ac[["year","aname","ID"]]
aGames=aGames1.set_axis(["year","hname","Count"], axis=1, inplace=False)
```

```python
Number_games=pd.merge(aGames,hGames , how="outer" , on=["year","hname"])

Number_games=Number_games.fillna(0)
Number_games
```

```python
Number_games["Count"]=Number_games["Count_x"]+Number_games["Count_y"]
Number_games.head()
```

```python
FinallMerged=pd.merge(Merged,Number_games , how="outer" , on=["year","hname"])
FinallMerged.head()
```

```python
FinallMerged["AVG hgoals"]= FinallMerged["hgoals"] / FinallMerged["Count"]
#FinallMerged["AVG agoals"]= FinallMerged["agoals"] / FinallMerged["Count"]
FinallMerged["AVG hPossesion"]= FinallMerged["hPossesion"] / FinallMerged["Count"]
#FinallMerged["AVG aPossesion"]= FinallMerged["aPossesion"] / FinallMerged["Count"]
FinallMerged["AVG hshotsOnTarget"]= FinallMerged["hshotsOnTarget"] / FinallMerged["Count"]
#FinallMerged["AVG ashotsOnTarget"]= FinallMerged["ashotsOnTarget"] / FinallMerged["Count"]
FinallMerged["AVG hshots"]= FinallMerged["hshots"] / FinallMerged["Count"]
#FinallMerged["AVG ashots"]= FinallMerged["ashots"] / FinallMerged["Count"]
FinallMerged["AVG hyellowCards"]= FinallMerged["hyellowCards"] / FinallMerged["Count"]
#FinallMerged["AVG ayellowCards"]= FinallMerged["ayellowCards"] / FinallMerged["Count"]
FinallMerged["AVG hredCards"]= FinallMerged["hredCards"] / FinallMerged["Count"]
#FinallMerged["AVG aredCards"]= FinallMerged["aredCards"] / FinallMerged["Count"]
FinallMerged["AVG hfouls"]= FinallMerged["hfouls"] / FinallMerged["Count"]
#FinallMerged["AVG afouls"]= FinallMerged["afouls"] / FinallMerged["Count"]
FinallMerged["AVG hsaves"]= FinallMerged["hsaves"] / FinallMerged["Count"]
#FinallMerged["AVG asaves"]= FinallMerged["asaves"] / FinallMerged["Count"]
FinallMerged["AVG h_rating"]= FinallMerged["h_rating"] / FinallMerged["Count"]
FinallMerged["AVG h_eoy_rank"]= FinallMerged["h_eoy_rank"] / FinallMerged["Count"]
FinallMerged["AVG h_boy_rank"]= FinallMerged["h_boy_rank"] / FinallMerged["Count"]
#FinallMerged["AVG a_rating"]= FinallMerged["a_rating"] / FinallMerged["Count"]
#FinallMerged["AVG a_eoy_rank"]= FinallMerged["a_eoy_rank"] / FinallMerged["Count"]
#FinallMerged["AVG a_boy_rank"]= FinallMerged["a_boy_rank"] / FinallMerged["Count"]
```

```python
FinallMerged["year_next"]=FinallMerged["year"]+1
```

```python
mix=pd.merge(Matches, FinallMerged ,how="left" ,right_on=["year_next","hname"] ,left_on=["year","hname"]  )
mix.head()
```

```python
mixwithaway=pd.merge(mix, FinallMerged ,how="left" ,right_on=["year_next","hname"] ,left_on=["year_x","aname"]  )
```

```python
mixwithaway.to_csv('/Users/macbook/Desktop/MIX.csv', index=False, header=True)
```

```python
Attributes2022=FinallMerged[FinallMerged.year_next==2022]
```

```python
from operator import itemgetter
```

```python
dfs = pd.read_html(r"https://en.wikipedia.org/wiki/2022_FIFA_World_Cup#Teams")
from collections.abc import Iterable

for i in range(len(dfs)):
    df = dfs[i]
    cols = list(df.columns.values)
    
    if isinstance(cols[0], Iterable):
        if any("Tie-breaking criteria" in c for c in cols):
            start_pos = i+1

        if any("Match 46" in c for c in cols):
            end_pos = i+1
matches = []
groups = ["A", "B", "C", "D", "E", "F", "G", "H"]
group_count = 0 

table = {}
#TABLE -> TEAM, POINTS, WIN PROBS (CRITERIO DE DESEMPATE)
table[groups[group_count]] = [[a.split(" ")[0], 0, []] for a in list(dfs[start_pos].iloc[:, 1].values)]

for i in range(start_pos+1, end_pos, 1):
    if len(dfs[i].columns) == 3:
        team_1 = dfs[i].columns.values[0]
        team_2 = dfs[i].columns.values[-1]
        
        matches.append((groups[group_count], team_1, team_2))
    else:
        group_count+=1
        table[groups[group_count]] = [[a, 0, []] for a in list(dfs[i].iloc[:, 1].values)]
```
