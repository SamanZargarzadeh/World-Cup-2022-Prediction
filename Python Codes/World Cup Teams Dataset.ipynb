```python
Matches=pd.read_csv(r'/Users/macbook/Desktop/Uni/Quarter 2/ML/World Cup Project/Matches.csv')
```

```python
WC_teams=["Ecuador", "Netherlands", "Qatar", "Senegal",
          "England", "Iran", "United States" , "Wales", 
          "Argentina", "Mexico", "Poland" , "Saudi Arabia" , 
          "Australia", "Denmark", "France" , "Tunisia",
          "Costa Rica", "Germany", "Japan" , "Spain",
          "Belgium", "Canada", "Croatia" , "Morocco" ,
          "Brazil", "Cameroon", "Serbia" , "Switzerland",
          "Ghana", "Portugal", "South Korea" , "Uruguay"]
```

```python
#Saman Method
Team_list=[]
WC_teams
for con in WC_teams:
    for i in range(len(Matches)):
        if ((Matches.home_team.iloc[i]==con) or (Matches.away_team.iloc[i]==con)):
            Team_list.append(Matches.iloc[i])
    Team=pd.DataFrame(Team_list)
    #Difference
    Team["d_Age"]=Team.home_age-Team.away_age
    Team["d_rank"]=Team.home_rank-Team.away_rank
    Team["d_OVA"]=Team.home_OVA-Team.away_OVA
    Team["d_ATT"]=Team.home_ATT-Team.away_ATT
    Team["d_MID"]=Team.home_MID-Team.away_MID
    Team["d_DEF"]=Team.home_DEF-Team.away_DEF
    Team["d_Goal"]=Team.home_score-Team.away_score
    
    for i in range(len(Team)):
        if (Team.away_team.iloc[i]==con):
            Team["d_Age"].iloc[i]=Team["d_Age"].iloc[i]*(-1)
            Team["d_rank"].iloc[i]=Team["d_rank"].iloc[i]*(-1)
            Team["d_OVA"].iloc[i]=Team["d_OVA"].iloc[i]*(-1)
            Team["d_ATT"].iloc[i]=Team["d_ATT"].iloc[i]*(-1)
            Team["d_MID"].iloc[i]=Team["d_MID"].iloc[i]*(-1)
            Team["d_DEF"].iloc[i]=Team["d_DEF"].iloc[i]*(-1)
            Team["d_Goal"].iloc[i]=Team["d_Goal"].iloc[i]*(-1)
            
            
    Team["Avg_goal"]=""
    for i in range(5,len(Team)):
        Team["Avg_goal"].iloc[i]=Team["d_Goal"].iloc[i-5:i].mean()

    Team["Class"]=""
    for i in range(len(Team)):
        if (Team["d_Goal"].iloc[i]>0):
            Team["Class"].iloc[i]="Win"
        elif (Team["d_Goal"].iloc[i]<0):
            Team["Class"].iloc[i]="Lose"
        else:
            Team["Class"].iloc[i]="Draw"
            
    Team=Team[Team.Year>2006]
  
    path='/Users/macbook/Desktop/Teams/'+str(con)+'.csv'
    Team.to_csv(path, index=False, header=True)
    Team_list=[]
    
```

```python
Matches["ID"]=""
for k in range(len(Matches)):
    Matches.ID.iloc[k]=k

Matches
```

```python
#Aneesh
AllTeam=pd.DataFrame()
Team_list=[]
WC_teams
for con in WC_teams:
    for i in range(len(Matches)):
        if ((Matches.home_team.iloc[i]==con) or (Matches.away_team.iloc[i]==con)):
            Team_list.append(Matches.iloc[i])
    Team=pd.DataFrame(Team_list)
    #Difference
    Team["d_Age"]=Team.home_age-Team.away_age
    Team["d_rank"]=Team.home_rank-Team.away_rank
    Team["d_OVA"]=Team.home_OVA-Team.away_OVA
    Team["d_ATT"]=Team.home_ATT-Team.away_ATT
    Team["d_MID"]=Team.home_MID-Team.away_MID
    Team["d_DEF"]=Team.home_DEF-Team.away_DEF
    Team["d_Goal"]=Team.home_score-Team.away_score
    
    for i in range(len(Team)):
        if (Team.away_team.iloc[i]==con):
            Team["d_Age"].iloc[i]=Team["d_Age"].iloc[i]*(-1)
            Team["d_rank"].iloc[i]=Team["d_rank"].iloc[i]*(-1)
            Team["d_OVA"].iloc[i]=Team["d_OVA"].iloc[i]*(-1)
            Team["d_ATT"].iloc[i]=Team["d_ATT"].iloc[i]*(-1)
            Team["d_MID"].iloc[i]=Team["d_MID"].iloc[i]*(-1)
            Team["d_DEF"].iloc[i]=Team["d_DEF"].iloc[i]*(-1)
            Team["d_Goal"].iloc[i]=Team["d_Goal"].iloc[i]*(-1)
            
            
    Team["home_avg_goal"]=""
    Team["away_avg_goal"]=""

    for i in range(5,len(Team)):
        if (Team.home_team.iloc[i]==con) :
            Team["home_avg_goal"].iloc[i]=Team["d_Goal"].iloc[i-5:i].mean()
        elif (Team.away_team.iloc[i]==con) :
            Team["away_avg_goal"].iloc[i]=Team["d_Goal"].iloc[i-5:i].mean()

    Team["Class"]=""
    for i in range(len(Team)):
        
        if (Team["d_Goal"].iloc[i]>0):
            if (Team.home_team.iloc[i]==con) :
                Team["Class"].iloc[i]="home Win"
            elif (Team.away_team.iloc[i]==con) :
                Team["Class"].iloc[i]="away Win"
        
        elif (Team["d_Goal"].iloc[i]<0):
            if (Team.home_team.iloc[i]==con) :
                Team["Class"].iloc[i]="away Win"
            elif (Team.away_team.iloc[i]==con) :
                Team["Class"].iloc[i]="home Win"        
        
        else:
            Team["Class"].iloc[i]="Draw"
            
    Team=Team[Team.Year>2006]
    AllTeam=pd.concat([AllTeam, Team] , axis=0, ignore_index=True)
    Team_list=[]

#END  
    
```

```python
AllTeamSorted=AllTeam.sort_values(by=['ID'] ,ascending=False)
AllTeamSorted.to_csv('/Users/macbook/Desktop/AneeshTeams/Allsort.csv', index=False, header=True)
```
