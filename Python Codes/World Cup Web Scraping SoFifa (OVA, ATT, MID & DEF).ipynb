```python
from bs4 import BeautifulSoup
import lxml
import requests
import pandas as pd
```

```python
url = "https://sofifa.com/teams?type=national&r=202200&set=true"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')
table= soup.find('table', class_="table table-hover persist-area")
table.find_all("th")
header=[]
for i in table.find_all("th"):
    title = i.text
    header.append(title)
for m in range(2,8):
    header[m]=header[m][:-1]

df=pd.DataFrame(columns = header)
table.find_all("tr")[1:]
for j in table.find_all("tr")[1:]:
    row_data= j.find_all("td")
    row = [tr.text for tr in row_data]
    length=len(df)
    df.loc[length]=row
    ind=0
for s in df.Name:
    df.Name[ind]=s[1:-14]
    ind+=1
WorldCup=df.iloc[:,1:]
WorldCup["Year"]=2023
WorldCup.to_csv(r'/Users/macbook/Desktop/SOFIFA_WorldCup.csv', index=False, header=True)
```
